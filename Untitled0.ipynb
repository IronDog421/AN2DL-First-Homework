```python
import numpy as np
import pandas as pd
import hashlib
import matplotlib.pyplot as plt

# Load the data
data = np.load('training_set.npz')
images = data['images']  # Original images
labels = data['labels']  # Corresponding labels

# Flatten the images for easy processing
images_flattened = images.reshape(images.shape[0], -1)

# Create a DataFrame to store images and labels
df = pd.DataFrame({'images': list(images_flattened), 'labels': labels.flatten()})

# Hash function to detect duplicates based on the image content
def image_hash(image_array):
    return hashlib.md5(image_array.tobytes()).hexdigest()

# Apply the hash function to each image and add a column for hashes
df['image_hash'] = df['images'].apply(lambda x: image_hash(np.array(x)))

# Drop duplicate images based on the hash, regardless of label
df_unique = df.drop_duplicates(subset=['image_hash'], keep='first').drop(columns=['image_hash'])

# Reset the index of the resulting DataFrame
df_unique = df_unique.reset_index(drop=True)

# Convert back to the original format if needed
unique_images = np.array(df_unique['images'].tolist()).reshape(-1, *images.shape[1:])
unique_labels = df_unique['labels'].to_numpy()

# Visualize random unique images and their labels
num_img = 20
random_indices = np.random.choice(len(unique_images), num_img, replace=False)

# Create subplots to display the images
fig, axes = plt.subplots(2, num_img // 2, figsize=(20, 9))

for i, idx in enumerate(random_indices):
    ax = axes[i // (num_img // 2), i % (num_img // 2)]
    ax.imshow(np.clip(unique_images[idx], 0, 255).astype(np.uint8))  # Ensure images are displayable
    ax.set_title(f"Label: {unique_labels[idx]}")

plt.tight_layout()
plt.show()

print(len(unique_images))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = np.load('training_set.npz')
images = data['images']  # Original images
labels = data['labels']  # Corresponding labels

# Flatten images for comparison purposes
images_flattened = images.reshape(images.shape[0], -1)

# Create DataFrame for easy manipulation
df = pd.DataFrame({'images': list(images_flattened), 'labels': labels.flatten()})

# Set to track unique images as tuples
unique_image_set = set()
unique_rows = []

# Iterate over rows to filter unique images
for _, row in df.iterrows():
    # Convert the flattened image to a tuple for easy comparison
    image_tuple = tuple(row['images'])
    if image_tuple not in unique_image_set:
        unique_image_set.add(image_tuple)
        unique_rows.append(row)

# Create a DataFrame from the unique rows
df_unique = pd.DataFrame(unique_rows).reset_index(drop=True)

# Convert back to original format if needed
unique_images = np.array(df_unique['images'].tolist()).reshape(-1, *images.shape[1:])
unique_labels = df_unique['labels'].to_numpy()

# Visualize a random selection of unique images and their labels
num_img = 20
random_indices = np.random.choice(len(unique_images), num_img, replace=False)

# Create subplots to display the images
fig, axes = plt.subplots(2, num_img // 2, figsize=(20, 9))

for i, idx in enumerate(random_indices):
    ax = axes[i // (num_img // 2), i % (num_img // 2)]
    ax.imshow(np.clip(unique_images[idx], 0, 255).astype(np.uint8))  # Ensure images are displayable
    ax.set_title(f"Label: {unique_labels[idx]}")
    ax.axis('off')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import cv2  # OpenCV library for image hashing
import matplotlib.pyplot as plt

# Load the data
data = np.load('training_set.npz')
images = data['images']  # Original images
labels = data['labels']  # Corresponding labels

# Reshape images if necessary
images_flattened = images.reshape(images.shape[0], -1)

# Convert to list format for DataFrame
df = pd.DataFrame({'images': list(images), 'labels': labels.flatten()})

# Function to compute perceptual hash for an image
def compute_phash(image_array):
    # Reshape image back to its original shape and convert to uint8 if needed
    image_reshaped = image_array.reshape(images.shape[1:]).astype(np.uint8)
    return cv2.img_hash.pHash(image_reshaped).tobytes()  # Get a hash as bytes

# Calculate the perceptual hash for each image and add it to a new column
df['image_hash'] = df['images'].apply(compute_phash)

# Remove duplicates by image_hash
df_unique = df.drop_duplicates(subset=['image_hash']).drop(columns=['image_hash']).reset_index(drop=True)

# Convert back to original format if needed
unique_images = np.array(df_unique['images'].tolist())
unique_labels = df_unique['labels'].to_numpy()

# Visualize a random selection of unique images and their labels
num_img = 20
random_indices = np.random.choice(len(unique_images), num_img, replace=False)

# Create subplots to display the images
fig, axes = plt.subplots(2, num_img // 2, figsize=(20, 9))

for i, idx in enumerate(random_indices):
    ax = axes[i // (num_img // 2), i % (num_img // 2)]
    ax.imshow(np.clip(unique_images[idx], 0, 255).astype(np.uint8))  # Display as uint8
    ax.set_title(f"Label: {unique_labels[idx]}")
    ax.axis('off')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import cv2  # OpenCV library for image hashing
import matplotlib.pyplot as plt

print(cv2.__version__)

# Load the data
data = np.load('training_set.npz')
images = data['images']  # Original images
labels = data['labels']  # Corresponding labels

# Reshape images if necessary
images_flattened = images.reshape(images.shape[0], -1)

# Convert to list format for DataFrame
df = pd.DataFrame({'images': list(images), 'labels': labels.flatten()})

# Function to compute perceptual hash for an image
def compute_phash(image_array):
    # Reshape image back to its original shape and convert to uint8 if needed
    image_reshaped = image_array.reshape(images.shape[1:]).astype(np.uint8)
    return cv2.img_hash.pHash(image_reshaped).tobytes()  # Get a hash as bytes

# Calculate the perceptual hash for each image and add it to a new column
df['image_hash'] = df['images'].apply(compute_phash)

# Separate duplicates before dropping them
duplicated_df = df[df.duplicated(subset=['image_hash'], keep='first')]

# Remove duplicates by image_hash
df_unique = df.drop_duplicates(subset=['image_hash']).drop(columns=['image_hash']).reset_index(drop=True)

# Convert back to original format for kept and dropped data
unique_images = np.array(df_unique['images'].tolist())
unique_labels = df_unique['labels'].to_numpy()

dropped_images = np.array(duplicated_df['images'].tolist())
dropped_labels = duplicated_df['labels'].to_numpy()

# Visualize retained images
num_img = 20
random_indices = np.random.choice(len(unique_images), min(num_img, len(unique_images)), replace=False)

fig, axes = plt.subplots(2, num_img // 2, figsize=(20, 9))
fig.suptitle("Kept Images", fontsize=16)

for i, idx in enumerate(random_indices):
    ax = axes[i // (num_img // 2), i % (num_img // 2)]
    ax.imshow(np.clip(unique_images[idx], 0, 255).astype(np.uint8))  # Display as uint8
    ax.set_title(f"Label: {unique_labels[idx]}")
    ax.axis('off')

plt.tight_layout()
plt.show()

# Visualize dropped images
if len(dropped_images) > 0:
    num_img_dropped = 20
    random_indices_dropped = np.random.choice(len(dropped_images), min(num_img_dropped, len(dropped_images)), replace=False)

    fig, axes = plt.subplots(2, num_img_dropped // 2, figsize=(20, 9))
    fig.suptitle("Dropped Images", fontsize=16)

    for i, idx in enumerate(random_indices_dropped):
        ax = axes[i // (num_img_dropped // 2), i % (num_img_dropped // 2)]
        ax.imshow(np.clip(dropped_images[idx], 0, 255).astype(np.uint8))  # Display as uint8
        ax.set_title(f"Label: {dropped_labels[idx]}")
        ax.axis('off')

    plt.tight_layout()
    plt.show()
else:
    print("No duplicate images were dropped.")

print(cv2.__version__)
```
